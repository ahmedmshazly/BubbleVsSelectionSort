
# Sorting Algorithms Performance Analysis

This notebook demonstrates the performance comparison between Bubble Sort and Selection Sort algorithms. We will implement both algorithms, measure their execution times for different input sizes, and plot these times to analyze their performance.


```python

import time
import matplotlib.pyplot as plt
import numpy as np
```

## Bubble Sort Implementation

```python

def bubbleSort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
```

## Selection Sort Implementation

```python

def selectionSort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[min_idx] > arr[j]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
```

## Function to Measure Execution Time

```python

def measure_time(sort_function, arr):
    start_time = time.time()
    sort_function(arr.copy())
    return time.time() - start_time
```

## Execution Time Measurement and Plotting

```python

input_sizes = np.arange(10, 501, 10)
bubble_sort_times = []
selection_sort_times = []

for size in input_sizes:
    arr = np.random.randint(0, 100000, size)
    bubble_sort_times.append(measure_time(bubbleSort, arr))
    selection_sort_times.append(measure_time(selectionSort, arr))

plt.figure(figsize=(10, 6))
plt.plot(input_sizes, bubble_sort_times, label='Bubble Sort', color='red')
plt.plot(input_sizes, selection_sort_times, label='Selection Sort', color='blue')
plt.xlabel('Input Size (Number of Elements)')
plt.ylabel('Execution Time (Seconds)')
plt.title('Bubble Sort vs Selection Sort Execution Time Comparison')
plt.legend()
plt.grid(True)
plt.show()
```
